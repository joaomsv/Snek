```python
def solution(A):
    A = [3, 2, 1, 1, 2, 3, 1]
    # A = [4, 1, 4, 1]
    # A = [2, 2, 2]
    # A = [1, 2, 3]
    # A = [1, 2, 3, 4, 4, 3, 2, 1]
    resultInc = 0
    resultDesc = 0
    A.sort()
    medianOdd = A[round(len(A)/2)]
    medianEven = A[round(len(A)/2) - 1]
    for num in A:
        resultInc += abs(num - medianOdd)
        resultDesc += abs(num - medianEven)
    return min(resultInc, resultDesc)

solution([])
```

```python
def calPoints(ops) -> int:
    result = None
    newOps = []
    for element in ops:
        if element.isdigit() or element[1::].isdigit():
            newOps.append(int(element))
        elif len(newOps) > 0 and element.upper() == 'C':
            newOps.pop()
        elif len(newOps) > 1 and element == '+':
            newOps.append(newOps[-1] + newOps[-2])
        elif len(newOps) > 0 and element.upper() == 'D':
            newOps.append(newOps[-1] * 2)

    result = sum(newOps)

    return result

# calPoints(['5', '2', 'c' , 'd', '+'])
calPoints(["5","-2","4","C","D","9","+","+"])
```

```python
def isValid(s: str) -> bool:
    result = None
    order = []
    for char in s:
        if char == '(':
            order.append(')')
        elif char == '[':
            order.append(']')
        elif char == '{':
            order.append('}')
        elif char == order[-1]:
            order.pop()
        else:
            break

    if not len(order):
        result = True
    else:
        result = False

    return result

# isValid('()[]{}')
# isValid('([])')
isValid('([)]')
```

```python
def solution(s: str) -> int:
    encoder = {'@': 1, '#': 5, '$': 10, '%': 50, '&': 100, '+': 500, '~': 1000}
    result = 0
    index = 0

    while index < len(s):
        char = s[index]
        if index < len(s)-1:
            if char in ('@') and s[index + 1] in ('#', '$') or char in ('$') and s[index + 1] in ('%', '&') or char in ('&') and s[index + 1] in ('+', '~'):
                result += (encoder[s[index + 1]] - encoder[char])
                index += 2
            else:
                result += encoder[char]
                index += 1
        else:
            result += encoder[char]
            index += 1

    return result

solution('$@@@')
# solution('@$')
# solution('~~&+$#')
```

```python
def print_rangoli(size):
    alphabet = [chr(x) for x in range(ord('a'), ord('z') + 1)]
    width = ((size * 2) - 1) + ((size - 1) * 2)

    for i in range(size * 2 - 1):
        line_size = (abs(size - abs(size - (i + 1))) * 2 - 1)
        line = [alphabet[abs(size - abs((line_size // 2) - abs((line_size // 2) - x) + 1))] for x in range(line_size)]
        line = '-'.join(line).center(width, '-')
        print(line)

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

```python
def minion_game(string: str):
    def calculate_game(originalstr: str, playerStr: 'list[str]') -> int:
        score = 0
        start = 0
        for char in playerStr:
            start = originalstr.find(char, start)
            score += len(originalstr[start:])
            start += 1

        return score

    vowels = ['A', 'E', 'I', 'O', 'U']
    stuart_score = 0
    kevin_score = 0
    # stuart = [x for x in string if x not in vowels]
    # kevin = [x for x in string if x in vowels]
    # stuartScore = calculate_game(string, stuart)
    # kevinScore = calculate_game(string, kevin)
    length = len(string)
    for i in range(length):
        if string[i] not in vowels:
            stuart_score += length - i
        elif string[i] in vowels:
            kevin_score += length - i

    if stuart_score > kevin_score:
        print(f'Stuart {stuart_score}')
    elif stuart_score < kevin_score:
        print(f'Kevin {kevin_score}')
    else:
        print(f'Draw')

minion_game(word)
```
